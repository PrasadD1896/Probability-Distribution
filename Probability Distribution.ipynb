# Probability Distribution

Probability theory is a branch of mathematics that deals with the study of random events and their associated probabilities. It is a fundamental tool in many fields, including statistics, economics, finance, physics, and engineering. In this article, we will explore the key concepts and techniques involved in probability theory.

## Basics of Probability:

Probability is a measure of the likelihood that a particular event will occur. It is expressed as a number between 0 and 1, where 0 represents an impossible event and 1 represents a certain event. For example, the probability of flipping a coin and it landing heads up is 0.5.

Probability can be expressed as a fraction, a decimal, or a percentage. The sum of the probabilities of all possible outcomes of an event is always equal to 1. For example, the probabilities of flipping a coin and it landing heads up or tails up are both 0.5, and their sum is 1.

```python
import numpy as np          #Importing Required Library
sample = np.random.random(10) #Generate a random sample of 10 numbers between 0 and 1
probability = sum(sample > 0.5) / len(sample) 
print("The probability of each number being greater than 0.5 is:", probability)
```

In this example, we used NumPy to generate a random sample and calculate the probability of each number being greater than 0.5. This is a basic example of working with probability in Python using the NumPy library.

## Probability Distributions

A probability distribution is a function that describes the probabilities of different outcomes of a random event.
There are two main types of probability distributions: 
1) Discrete probability distributions
    2) Continuous probability distributions.

### 1) Discrete Probability Distribution

Discrete probability distributions are used to describe events that can only take on a finite number of outcomes. The      probabilities of each outcome are given by the probability mass function (PMF), which assigns a probability to each possible outcome. 
Examples of discrete probability distributions include the binomial distribution, the Poisson distribution, and the hypergeometric distribution.

In Python, there are several libraries that provide functions for working with discrete probability distributions. One of the most commonly used libraries for working with probability distributions in Python is SciPy. Let's take a look at how to use SciPy to work with discrete probability distributions in Python.

#### 1.1 Binomial Probability Distribution

Binomial probability distribution is a discrete probability distribution that is commonly used in statistical analysis to model the number of successes in a fixed number of independent trials. It is used to calculate the probability of obtaining a certain number of successes in a fixed number of trials, when each trial has only two possible outcomes (success or failure) and the probability of success is the same for each trial.

The binomial distribution is defined by two parameters: n, the number of trials, and p, the probability of success in each trial. The probability mass function (PMF) of the binomial distribution gives the probability of obtaining exactly k successes in n trials, and can be calculated using the formula:

P(X = k) = (n choose k) * p^k * (1 - p)^(n - k)

where "choose" represents the binomial coefficient.

The mean and variance of the binomial distribution are given by:

μ = n * p

σ^2 = n * p * (1 - p)

The binomial distribution is used in a variety of applications, including quality control, survey sampling, and hypothesis testing. In Python, the binomial distribution can be easily modeled and analyzed using the SciPy library.

```python
from scipy.stats import binom    #Importing Binomial Distribution from Scipy
n = 10                           #The n parameter is the number of trials 
p = 0.5                          #The p parameter is the probability of success
binom_dist = binom(n, p)         #binom() function from SciPy to create a binomial distribution 
pmf = binom_dist.pmf(range(n+1)) #pmf() function calculate the PMF of the binomial distribution
cdf = binom_dist.cdf(range(n+1)) #cdf() function to calculate the CDF

import matplotlib.pyplot as plt  #plot the PMF and CDF using Matplotlib
plt.plot(range(n+1), pmf, 'bo', label='PMF') 
plt.plot(range(n+1), cdf, 'ro', label='CDF')
plt.legend()
plt.show()
```

The resulting plot should show the PMF and CDF of the binomial distribution. This is a basic example of working with discrete probability distributions in Python using the SciPy and Matplotlib libraries.

#### 1.2 Poisson Distribution 

Poisson distribution is a discrete probability distribution that is commonly used in statistical analysis to model the number of events that occur in a fixed interval of time or space, given the average rate at which events occur. It is used to calculate the probability of a certain number of events occurring in a given interval, when the events occur randomly and independently of each other.

The Poisson distribution is defined by a single parameter, λ (lambda), which represents the mean number of events in the interval. The probability mass function (PMF) of the Poisson distribution gives the probability of obtaining exactly k events in the interval, and can be calculated using the formula:

P(X = k) = e^(-λ) * (λ^k) / k!

where e is the mathematical constant e (approximately 2.71828), and k! represents the factorial of k.

The mean and variance of the Poisson distribution are both equal to λ.

The Poisson distribution is used in a variety of applications, including modeling rare events such as accidents or failures, estimating call center traffic, and modeling the number of mutations in a DNA sequence. In Python, the Poisson distribution can be easily modeled and analyzed using the SciPy library.

```python
from scipy.stats import poisson    #Importing Poisson Distribution from Scipy
lam = 3                            #The lambda parameter is the mean of the distribution
poisson_dist = poisson(lam)        #poisson() function from SciPy to create a Poisson distribution object 
pmf = poisson_dist.pmf(range(11))  #pmf() function calculate the PMF of the poisson distribution
cdf = poisson_dist.cdf(range(11))  #cdf() function to calculate the CDF

import matplotlib.pyplot as plt    #plot the PMF and CDF using Matplotlib
plt.plot(range(11), pmf, 'bo', label='PMF')
plt.plot(range(11), cdf, 'ro', label='CDF')
plt.legend()
plt.show()
```

The resulting plot should show the PMF and CDF of the Poisson distribution. This is a basic example of working with the Poisson distribution in Python using the SciPy and Matplotlib libraries.

#### 1.3 Hypergeometric Distribution

Hypergeometric distribution is a discrete probability distribution that is commonly used in statistical analysis to model situations where a sample is drawn from a finite population without replacement. It is used to calculate the probability of obtaining a certain number of elements of interest (successes) in a sample of fixed size, drawn from a population of fixed size that contains a known number of elements of interest.

The Hypergeometric distribution is defined by three parameters: N, the size of the population, M, the number of elements of interest in the population, and n, the sample size. The probability mass function (PMF) of the Hypergeometric distribution gives the probability of obtaining exactly k successes in a sample of size n, and can be calculated using the formula:

P(X = k) = (M choose k) * (N - M choose n - k) / (N choose n)

where "choose" represents the binomial coefficient.

The mean and variance of the Hypergeometric distribution are given by:

μ = n * M / N

σ^2 = n * M * (N - M) * (N - n) / (N^2 * (N - 1))

The Hypergeometric distribution is used in a variety of applications, including quality control, population genetics, and sampling. In Python, the Hypergeometric distribution can be easily modeled and analyzed using the SciPy library.

```python
from scipy.stats import hypergeom    #Importing Hypergeometrical Distribution from Scipy
N = 20                               #population size=20
M = 8                                #number of success states in the population
n = 5                                #number of draws=5
hypergeom_dist = hypergeom(N, M, n)  #hypergeom() function from SciPy to create a Hypergeometric distribution object
pmf = hypergeom_dist.pmf(range(6))   #pmf() function calculate the PMF of the hypergeometrical distribution
cdf = hypergeom_dist.cdf(range(6))   #cdf() function calculate the CDF of the hypergeometrical distribution

import matplotlib.pyplot as plt      #plot the PMF and CDF using Matplotlib
plt.plot(range(6), pmf, 'bo', label='PMF')
plt.plot(range(6), cdf, 'ro', label='CDF')
plt.legend()
plt.show()
```

The resulting plot should show the PMF and CDF of the Hypergeometric distribution. This is a basic example of working with the Hypergeometric distribution in Python using the SciPy and Matplotlib libraries.

### 2) Continous Probability Distribution

Continuous probability distributions are used to describe events that can take on any value within a range. The probabilities of different values are given by the probability density function (PDF), which gives the probability of a random variable falling within a particular range of values. Examples of continuous probability distributions include the normal distribution, the exponential distribution, and the gamma distribution.

#### 2.1 Normal Distribution / Gaussian Distribution

Normal distribution, also known as Gaussian distribution, is a continuous probability distribution that is commonly used in statistical analysis to model the distribution of a random variable that has a large number of possible values. It is a bell-shaped curve that is symmetric around its mean value and characterized by two parameters: the mean (μ) and the standard deviation (σ).

The normal distribution has several important properties that make it useful in modeling real-world phenomena. Firstly, many natural phenomena follow a normal distribution, including the heights and weights of individuals in a population, scores on standardized tests, and errors in measurements. Secondly, the normal distribution is a central limit theorem (CLT) distribution, which means that the sum of independent and identically distributed (i.i.d.) random variables tends to follow a normal distribution as the number of variables increases.

The probability density function (PDF) of the normal distribution is given by the formula:

f(x) = (1 / (σ * sqrt(2π))) * exp(-((x - μ)^2 / (2 * σ^2)))

where exp represents the exponential function, and sqrt represents the square root function.

The mean, median, and mode of the normal distribution are all equal to μ. About 68% of the data falls within one standard deviation (σ) of the mean, about 95% falls within two standard deviations, and about 99.7% falls within three standard deviations.

The normal distribution is widely used in statistical analysis, including hypothesis testing, confidence intervals, and regression analysis. In Python, the normal distribution can be easily modeled and analyzed using the SciPy library.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Define the mean and standard deviation of the normal distribution
mu = 0
sigma = 1

# Generate a random sample from the normal distribution
sample = np.random.normal(mu, sigma, size=1000)

# Calculate the probability density function (PDF) for the sample
pdf = norm.pdf(sample, mu, sigma)

# Plot the histogram of the sample and overlay the PDF
plt.hist(sample, bins=30, density=True, alpha=0.5)
plt.plot(sample, pdf, color='red')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Example')
plt.show()
```

The resulting plot shows the bell-shaped curve of the normal distribution.

#### 2.2 Exponential Distribution

Exponential distribution is a continuous probability distribution that is commonly used in statistical analysis to model the time between events that occur at a constant rate, also known as the memoryless property. It is a key distribution in survival analysis, queuing theory, and reliability analysis.

The probability density function (PDF) of the exponential distribution is given by the formula:

f(x) = λ * exp(-λ * x)

where λ is the rate parameter, and exp represents the exponential function.

The cumulative distribution function (CDF) of the exponential distribution is given by the formula:

F(x) = 1 - exp(-λ * x)

where F represents the CDF.

The mean and standard deviation of the exponential distribution are both equal to 1/λ.

The exponential distribution is widely used in statistical analysis, including reliability analysis, queuing theory, and survival analysis. In Python, the exponential distribution can be easily modeled and analyzed using the SciPy library.

```python
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt

# Define the rate parameter of the exponential distribution
lam = 0.5

# Generate a random sample from the exponential distribution
sample = np.random.exponential(scale=1/lam, size=1000)

# Calculate the probability density function (PDF) for the sample
pdf = expon.pdf(sample, scale=1/lam)

# Plot the histogram of the sample and overlay the PDF
plt.hist(sample, bins=30, density=True, alpha=0.5)
plt.plot(sample, pdf, color='red')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution Example')
plt.show()
```

The resulting plot shows the decreasing exponential curve of the exponential distribution.

#### 2.3 Gamma Distribution

The gamma distribution is a continuous probability distribution that is widely used in statistical analysis to model a variety of different types of data. It is a two-parameter family of distributions that can be used to model data with a wide range of shapes, including skewed, symmetric, and bimodal data.

The probability density function (PDF) of the gamma distribution is given by the formula:

f(x) = (x^(k-1) * exp(-x/theta)) / (theta^k * Gamma(k))

where x is the random variable, k is the shape parameter, theta is the scale parameter, and Gamma(k) is the gamma function.

The mean and variance of the gamma distribution are given by:

mean = k * theta
variance = k * theta^2

The gamma distribution is widely used in many different areas of statistics and applied mathematics, including finance, physics, and engineering. It can be used to model the lifetime of a product, the time between two events, the number of arrivals in a fixed interval of time, and much more.

In Python, the gamma distribution can be easily modeled and analyzed using the SciPy library, which provides a wide range of functions for working with the gamma distribution.

```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt

# Define the shape and scale parameters
shape = 2
scale = 1

# Generate a random sample from the gamma distribution
sample = np.random.gamma(shape, scale, size=1000)

# Calculate the probability density function (PDF) for the sample
pdf = gamma.pdf(sample, shape, scale)

# Plot the histogram of the sample and overlay the PDF
plt.hist(sample, bins=30, density=True, alpha=0.5)
plt.plot(sample, pdf, color='red')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Gamma Distribution Example')
plt.show()
```

The resulting plot shows the characteristic shape of the gamma distribution.

## Applications of Probability Distribution:

Probability distributions have a wide range of applications in various fields, including finance, engineering, physics, and biology. They are used to model and analyze various types of data, such as stock prices, failure rates, radioactive decay, and enzyme activity.

In finance, probability distributions are used to model and analyze the behavior of financial assets, such as stocks and bonds. The normal distribution is commonly used to model the returns of a portfolio of stocks, while the Poisson distribution is used to model the frequency of defaults on bonds.

In engineering, probability distributions are used to model and analyze the failure rates of mechanical and electronic systems. The exponential distribution is commonly used to model the time to failure of a system, while the Weibull distribution is used to model the failure rate of a system over time.

In physics, probability distributions are used to model and analyze various phenomena, such as the decay of radioactive isotopes and the distribution of particle energies. The exponential and gamma distributions are commonly used to model radioactive decay, while the normal distribution is used to model the distribution of particle energies.

In biology, probability distributions are used to model and analyze various biological processes, such as enzyme activity and gene expression. The Poisson distribution is commonly used to model the number of enzyme-catalyzed reactions per unit time, while the gamma distribution is used to model the time until a specified number of reactions occur.
